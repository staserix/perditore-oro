```python
import pandas as pd

# Carica il file CSV
df = pd.read_csv('/content/GAU_EUR Dati Storici.csv')

# Seleziona solo le colonne "Data" e "Ultimo"
df_subset = df[['Data', 'Ultimo']].copy()

# Converti la colonna 'Data' in formato datetime
df_subset['Data'] = pd.to_datetime(df_subset['Data'])

# Imposta la colonna 'Data' come indice del dataframe
df_subset = df_subset.set_index('Data')

# Converti la colonna 'Ultimo' in numerico, gestendo il separatore decimale con la virgola
# Prima sostituisci la virgola con il punto, poi converti in float
df_subset['Ultimo'] = df_subset['Ultimo'].astype(str).str.replace(',', '.', regex=False).astype(float)


# Creazione di lag features (valori precedenti)
df_subset['Ultimo_lag1'] = df_subset['Ultimo'].shift(1)
df_subset['Ultimo_lag7'] = df_subset['Ultimo'].shift(7)

# Creazione di moving average
df_subset['Ultimo_rolling_mean7'] = df_subset['Ultimo'].rolling(window=7).mean()

# Elimina le righe con valori NaN introdotti dalle lag e moving average
df_subset = df_subset.dropna()

# Mostra le prime righe del dataframe con le nuove features
print(df_subset.head())

# Definisci la variabile target (y)
y = df_subset['Ultimo']

# Definisci le feature (X) - escludi la colonna target stessa
X = df_subset.drop('Ultimo', axis=1)

# Mostra le prime righe di X e y
print("Features (X):")
print(X.head())
print("\nTarget (y):")
print(y.head())
```

```python
from sklearn.model_selection import train_test_split

# Poiché è una serie temporale, non usiamo train_test_split casuale
# Dividiamo i dati in base a una data o una percentuale
# Supponiamo di voler usare l'80% dei dati per l'addestramento e il 20% per il test

train_size = int(len(df_subset) * 0.8)

X_train, X_test = X[0:train_size], X[train_size:len(df_subset)]
y_train, y_test = y[0:train_size], y[train_size:len(df_subset)]

print(f"Dimensioni del set di addestramento: X_train = {X_train.shape}, y_train = {y_train.shape}")
print(f"Dimensioni del set di test: X_test = {X_test.shape}, y_test = {y_test.shape}")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Inizializza il modello di Regressione Lineare
model = LinearRegression()

# Addestra il modello sui dati di addestramento
model.fit(X_train, y_train)

# Fai previsioni sul set di test
y_pred = model.predict(X_test)

# Mostra le prime previsioni
print("Prime previsioni:")
print(y_pred[:10])
```

```python
# Valuta il modello sul set di test
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"\nErrore Quadratico Medio (MSE) sul set di test: {mse:.4f}")
print(f"Coefficiente di Determinazione (R²) sul set di test: {r2:.4f}")
```

```python
import matplotlib.pyplot as plt

# Crea un grafico per confrontare i valori reali e le previsioni sul set di test
plt.figure(figsize=(12, 6))
plt.plot(y_test.index, y_test, label='Valori Reali')
plt.plot(y_test.index, y_pred, label='Previsioni')
plt.title('Previsioni vs Valori Reali (Set di Test)')
plt.xlabel('Data')
plt.ylabel('Valore Ultimo')
plt.legend()
plt.grid(True)
plt.show()
```
